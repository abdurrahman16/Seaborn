```python
import seaborn as sns
```

```python
dataset_iris=sns.load_dataset('iris')
```

```python
dataset_iris.head()
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=dataset_iris, x='sepal_length', y='sepal_width' ,palette='Set2', hue='species')
plt.title('Iris Species')
plt.show()
```

```python
sns.histplot(data=dataset_iris, x='sepal_length', y='sepal_width' ,palette='Set2', hue='species')
```

```python
fmri= sns.load_dataset('fmri')
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal',  style='region',hue='event', markers=True, dashes=False)
```
